### 構文（if文）
```Python
if <条件>:
    <文１>
    <文２>
    ・・・
```

### Sample1.py

````python
sale = int(input("売上を入力してください。"))

if sale >= 100 :
    print("売上は好調です。")

print("処理を終了します。")
````

### Sample2.py

````python
sale = int(input("売上を入力してください。"))

if sale >= 100:
    print("売上は好調です。")
elif sale >= 50:
    print("売上は普通です。")
else:
    print("売上は低調です。")

print("処理を終了します。")
````

### Sample3.py

````python
sale = int(input("売上を入力してください。"))
num = int(input("人数を入力してください。"))

if sale >= 100 and num >= 30 :
    print("売上は大盛況です。")
elif sale >= 100:
    print("売上は好調です。")
elif sale >= 50:
    print("売上は普通です。")
else:
    print("売上は低調です。")

print("処理を終了します。")
````

### 構文（for文）
```Python
for <変数> in <繰り返し反復処理できる仕組み>:
    ・・・
```

### Sample4.py

````python
for i in range(12):
    print(i+1 , "月のデータです。")
````

### 構文（for文）
```Python
while <条件>:
    ・・・
```

### Sample5.py

````python
i = 1

while i <= 12:
    print(i , "月のデータです。")
    i = i+1
````

### Sample6.py

````python
for i in range(5):
    for j in range(3):
        print("iは", i, ":jは", j)
````

### Sample7.py

````python
v = False

for i in range(5):
    for j in range(5):
        if v is False:
            print("*", end="")
            v = True
        else:
            print("-", end="")
            v = False
    print()
````

### 構文（break文）
```Python
break
```

### Sample8.py

````python
num = int(input("何月の処理を飛ばしますか？(1～12)"))

for i in range(12):
    if (i+1) == num:
        continue
    print(i+1 , "月のデータです。")
````

### 構文（continue文）
```Python
continue
```

### Sample9.py

````python
num = int(input("何月の処理で終了しますか？(1～12)"))

for i in range(12):
    print(i+1 , "月のデータです。")
    if (i+1) == num:
        break
````
