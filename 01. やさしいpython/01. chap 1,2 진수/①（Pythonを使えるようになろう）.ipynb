````python
#問題3
#forループとif文をつかってy=x/3のグラフっぽいやつ書いてみよう
#10×10より大きいサイズで
#どこ書き換えてもいいよ

for y in range(20):
    for x in range(40):
        if(y==x*3):
            print("●",end="")
        else:
            print("〇",end="")
    print()
````
