```python
# writerow()やwriterows()を使って、九九表をganbare.csvに出力してみてね
#printしたときに、変な空行が入らないようにしてね！

'''
出力例
[1,2,...,8,9]
[2,4,...,16,18]
...
[8,16,...,64,72]
[9,18,...,72,81]
'''

import csv


# ここにかいてね
import numpy as np

ganbare=np.empty([9,9], dtype='int')

for i in range(1,10):
    for y in range(1,10):
        ganbare[i-1][y-1]=str(i*y)
        
g = open("ganbare.csv", "w",newline="")
w = csv.writer(g)
w.writerows(ganbare)
g.close()

with open('ganbare.csv') as f:
    reader = csv.reader(f)
    l = [row for row in reader]
    for r in l:
        print(r)
```

```python
#numpyの二次元配列cを左右反転して、全部５倍にしてね
# 二行が理想、importより下３行が理解できたらめっちゃいい
'''
例
c= [[1,2,3]
    [4,5,6]
    [7,8,9]]
なら
res=[[15,10,5]
    [30,25,20]
    [45,40,35]]
'''
import numpy as np
a=np.array([1,2,3,4,5])
a=a.reshape([5,1])
b=-np.array([1,2,3,4,5])

print("-----a------")
print(a)

print("-----b------")
print(b)

c=a+b
print("-----c------")
print(c)

# ここにかいてね
res=np.flip(c*5,axis=1)
print("-----出力-----")
print(res)
```
