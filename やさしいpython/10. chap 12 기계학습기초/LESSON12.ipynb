```python
import matplotlib.pyplot as plt

data = [8, 17, 0, 11, 6, 21, 16, 6, 17, 11, 
        7, 9, 6, 13, 12, 16, 3, 14, 13, 12]

plt.title("Histogram")

plt.xlabel("value")
plt.ylabel("frequency")

plt.hist(data, bins=8, color='m')
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

x = []
y = []

for i in range(100):
    x.append(random.uniform(0, 50))
    y.append(random.uniform(0, 50))
    
plt.xlim(-100,100)
plt.ylim(-100,100)
plt.scatter(x, y, marker="x")
plt.show()
```

```python
import matplotlib.pyplot as plt

data2 = []
for i in range(1000):
    data2.append(random.normalvariate(0,10))

plt.title("Histogram")
plt.hist(data2, bins=50)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-8,8,0.01)
y = 3*x+5
y2 = x*x

plt.title("y=f(x)")
plt.xlabel("x")
plt.ylabel("y")

plt.plot(x, y, x, y2)
plt.grid(True)
plt.show()
```
