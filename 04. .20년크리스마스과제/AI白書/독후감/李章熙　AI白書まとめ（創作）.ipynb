## ですから、 GANについて調査しました。

まず、Generativeは生成モデルということを意味します。<br>
生成モデルとは「もっともらしい偽物」を生み出します。 実在しそうな人の顔や、猫などの写真を作り上げることを意味します。 ここで「もっともらしい」という現実に存在しそうな人、例えば身長176cmに体重75kg程度なら現実でもよく見られる数値です。 でも、198センチで45キロというなら存在するかもしれませんが、データの数値の上でもほとんどない組み合わせです。 つまり、数学的に見ると、実際のデータの分布に似た分布のデータを意味します。<br>
単純に身長と体重の生成モデルを作ることは比較的簡単なことです。 しかし、人の顔の生成モデルを作るためには、写真の横ピクセルx縦ピクセルxRBG個の変数が発生するので難しいことがあります。<br>

Adversarialは二つのモデルを敵対的(Adversarial)に競争させるということを意味します。<br>
最も有名な例として、偽札犯と警察があります。 当然この二人は互いに敵対的な競争関係です。 偽造紙幣犯は警察に捕まらないために、ますます偽造紙幣の製造技術を発展させ、警察は偽造紙幣犯を捕えるために検査技術を発展させるでしょう。 このようにして時間の経過とともに、偽造紙幣犯の偽造技術は完璧に近く発展するでしょう。(もちろん現実ではその前に銀行で偽造防止デザインを変えるでしょうけど)<br>
このように、偽造紙幣犯に該当する生成者(Generator)と警察に該当する区分者(Discriminator)を競争させて学習させます。<br>
つまり、生成者がもっともらしい偽物を作って区分者をだまし、区分者は生成者が作ったデータと本物を区別することを繰り返し、本物に近い本物と区別できない偽物を作ることができる生成者を獲得します。<br>




GAN은 Generative Adversarial Network의 약자 입니다.<br>

먼저 Generative는 생성 모델 이라는것을 의미 합니다. <br>
생성 모델이란 '그럴듯한 가짜'를 만들어냅니다. 실재로 있을법한 사람의 얼굴이나, 고양이등의 사진을 만들어 내는것을 의미 합니다. 여기서 '그럴듯하다'는 현실에 존재할것 같은 사람, 예를 들면 키 176cm에 몸무게 75kg 정도면 현실에서도 자주 볼수 있는 수치 입니다. 하지만 198cm에 몸무게 45kg 이라고 한다면 존재할지도 모르겠지만 데이터 수치상으로도 거의 없는 조합입니다. 즉 수학적으 본다면 실제 데이터의 분포와 비슷한 분포에서 나온 데이터를 의미 합니다.<br>
단순하게 키와 몸무게의 생성 모델을 만드는것은 비교적 쉬운 일입니다. 하지만 사람의 얼굴 생성 모델을 만들려면 사진의 가로픽셀 x 세로픽셀 x RBG 개의 변수가 생기므로 어려운 일이 됩니다. <br>

Adversarial는 두개의 모델을 적대적(Adversarial)으로 경쟁 시킨다는것을 의미 합니다.
가장 유명한 예로 위조지폐범과 경찰이 있습니다. 당연히 이 둘은 서로 적대적인 경쟁 관계 입니다. 위조지폐범은 경찰에게 잡히지 않기 위해서 점점 위조지폐 제조 기술을 발전 시킬 것이고, 경찰은 위조지폐범을 잡기 위해 검사 기술을 발전 시킬 것입니다. 이렇게 시간이 흐르면서 위조지폐범의 위조기술은 완벽에 가까워 지게 발전할 것입니다.(물론 현실에서는 그전에 은행에서 위조방지 디자인을 바꾸겠지만ㅋ)<br>
이렇듯이 위조지폐범에 해당하는 생성자(Generator)와 경찰에 해당하는 구분자(Discriminator)를 경쟁시켜 학습을 시킵니다.<br>
즉 생성자가 그럴듯한 가짜를 만들어 구분자를 속이고, 구분자는 생성자가 만든 데이터와 진짜를 구별하는것을 반복하여 진짜에 가까운, 진짜와 구별할 수 없는 가짜를 만들수 있는 생성자를 얻어 냅니다.<br>

### GANの限界<br>

GANの最大の問題点は学習させるのが難しいという点です。 学習が上手く行われるためには、似たようなレベルで一緒に発展しなければなりませんが、一方が早く成長してしまうと競争関係が終わってしまいます。 つまり警察が強力になりすぎて、偽札犯がつかまって誰も偽札を作らなくなってしまうということです。<br>
また、モード崩壊(Mode Collapse)という現象も起きたりします。 生成者が様々なものではなく、似たようなものだけを作り出す場合を意味します。<br>

他の限界点は音声やイメージの分野では多くの成功事例が(上に出てくる放送でも歌う歌手のイメージと歌を歌う音声の部分で使用しました。)ありましたが、まだテキストを生成する部分ではまだ成功した事例がほとんどないということです。<br>

GAN의 한계<br>

GAN의 가장큰 문제점은 학습시키기가 어렵다는 점 입니다. 학습이 잘 이루어지기 위해서는 서로 비슷한 수준으로 함께 발전이 되어야 하는데 한쪽이 너무 빠르게 성장을 해버리면 경쟁관계가 끝나버린다는 것입니다. 즉 경찰이 너무 강력해져서 위조지폐범들이 다 잡혀 들어가서 아무도 위조 지폐를 만들지 않게 되어 버리는 것 입니다.<br>
그리고 모드붕괴(Mode Collapse)라는 현상도 일어나기도 합니다. 생성자가 다양한 것이 아닌 비슷한것만 만들어 내는 경우를 의미 합니다.<br>

다른 한계점은 음성이나 이미지 분야에서는 많은 성공 사례들이(위에 나오는 방송에서도 노래하는 가수의 이미지와 노래를 부르는 음성 부분에서 사용 하였습니다.) 있었지만, 아직 텍스트를 생성하는 부분에서는 아직 성공한 사례가 거의 없다는 것입니다.<br>
