# Chapter11　matplotlibの使い方

## 11.1 1種類のデータを可視化する

### 11.1.1 グラフにデータをプロットする
* matplotlibを用いたグラフの作成
    * **`matplotlib.pyplot.plot(X軸に対応するデータ, Y軸に対応するデータ)`** でグラフを生成
    * **`matplotlib.pyplot.show()`** で画面に表示
    * Jupyter Notebookの場合、グラフ表示前に一度 **`%matplotlib inline`** と記述する
        * 記述しないと（1回目の実行時に）グラフが表示されない

#### グラフ作成の例

```python
import matplotlib.pyplot as plt
import numpy as np
## Jupyter Notebookの場合必要
%matplotlib inline

## X軸データ
x = np.linspace(-10, 10)   ## -10から10まで等間隔のデータ取得（50個）
## Y軸データ
y = 0.5 * x + 10

# データx,yをグラフにプロット
plt.plot(x, y)

# グラフ表示
plt.show()
```

### 11.1.2 グラフの表示範囲を設定する
* matplotlib.pyplotによるグラフ作成時、**表示範囲は自動的にすべてのデータが可視化される**
    * 表示範囲はデータの **min(), max()** となる
* データの一部を表示したい場合は範囲指定を行うことができる
    * X軸の範囲指定： **`matplotlib.pyplot.xlim([最小値, 最大値])`**
    * Y軸の範囲指定： **`matplotlib.pyplot.ylim([最小値, 最大値])`**

#### グラフ作成の例

```python
import matplotlib.pyplot as plt
import numpy as np
## Jupyter Notebookの場合必要
%matplotlib inline

## X軸データ
x = np.linspace(-10, 10)   ## -10から10まで等間隔のデータ取得（50個）
## Y軸データ
y = 0.5 * x + 10

## 表示範囲指定（X軸）
plt.xlim(-5, 5)

# データx,yをグラフにプロット
plt.plot(x, y)

# グラフ表示
plt.show()
```

### 11.1.3 グラフの要素に名前を設定する
* グラフタイトルの設定
    * **`matplot.pyplot.title("タイトル")`**
* X軸の名前
    * **`matplot.pyplot.xlabel("X軸の名前")`**
* Y軸の名前
    * **`matplot.pyplot.ylabel("Y軸の名前")`**


#### グラフタイトル、X軸・Y軸設定の例

```python
import matplotlib.pyplot as plt
import numpy as np
## Jupyter Notebookの場合必要
%matplotlib inline

## X軸データ
x = np.linspace(-10, 10)   ## -10から10まで等間隔のデータ取得（50個）
## Y軸データ
y = 0.5 * x + 10

## グラフタイトル設定
plt.title("y = 0.5 * x + 10")
## X軸設定
plt.xlabel("X")
## Y軸設定
plt.ylabel("Y")

# データx,yをグラフにプロット
plt.plot(x, y)

# グラフ表示
plt.show()
```

### 11.1.4 グラフにグリッドを表示する
* グリッド（格子）表示はデフォルト非表示となっている
* グリッド表示： **`matplotlib.pyplot.grid(True)`**

#### グリッド表示の例

```python
import matplotlib.pyplot as plt
import numpy as np
## Jupyter Notebookの場合必要
%matplotlib inline

## X軸データ
x = np.linspace(-10, 10)   ## -10から10まで等間隔のデータ取得（50個）
## Y軸データ
y = 0.5 * x + 10

## グリッド表示設定
plt.grid(True)

# データx,yをグラフにプロット
plt.plot(x, y)

# グラフ表示
plt.show()
```

### 11.1.5 グラフの軸に目盛りを設定する
* グラフ作成時、X軸・Y軸は自動的に目盛りがつく
* 目盛りを手動で設定する場合
    * X軸： **`matplotlib.pyplot.xticks(目盛りを挿入する位置, 挿入する目盛り)`**
    * Y軸： **`matplotlib.pyplot.yticks(目盛りを挿入する位置, 挿入する目盛り)`**
    * 挿入する目盛りが複数ある場合、位置・目盛りはそれぞれリストにする

#### グラフの軸に目盛りを設定する例

```python
import matplotlib.pyplot as plt
import numpy as np
## Jupyter Notebookの場合必要
%matplotlib inline

## X軸データ
x = np.linspace(-10, 10)   ## -10から10まで等間隔のデータ取得（50個）
## Y軸データ
y = 0.5 * x + 10

## グリッド表示設定
plt.grid(True)

## X軸の目盛りを設定する
xpos = [-8.0, -4.0, 0, 2.0, 3.0, 5.0, 8.0]
xlabel = ["▲8", "▲4", "0", "2", "3", "5", "8"]
plt.xticks(xpos, xlabel)

# データx,yをグラフにプロットし表示
plt.plot(x, y)

# グラフ表示
plt.show()
```

## ●11.2 複数のデータを可視化する①

### 〇11.2.1  1つのグラフに2種類のデータをプロットする
* 1つのグラフに複数のデータをプロットする
    * **`matplotlib.pyplot.plot()`** を異なるデータで繰り返し呼ぶことでプロット可能
* 複数のデータがグラフ上で区別できるよう、プロットの色の指定が可能
    * **`matplotlib.pyplot.plot(x, y, color="色指定")`**
        * colorに指定する色は **HTMLカラーコード** で指定する
            * **#(16進数6桁)** で指定
            * googleで「色見本」と検索すると各色に対応するカラーコードの検索が可能
        * 代表的な色は文字での指定も可能
            * "b"：青、"g"：緑、"r"：赤、"c"：シアン、"m"：マゼンダ、"y"：黄色、"k"：黒、"w"：白

#### 1つのグラフに2種類のデータをプロットする例

```python
import matplotlib.pyplot as plt
import numpy as np
## Jupyter Notebookの場合必要
%matplotlib inline

## X軸データ
x = np.linspace(-10, 10)   ## -10から10まで等間隔のデータ取得（50個）
## Y軸データ1
y1 = 0.5 * x + 10
## Y軸データ2
y2 = -0.5 * x + 15

# グリッド表示設定
plt.grid(True)

# データx,y1をグラフにプロット
# HTMLカラーコードで赤（#aa0000）を指定
plt.plot(x, y1, color="#aa0000")

# データx,y2をグラフにプロット
# 文字指定で緑を指定
plt.plot(x, y2, color="g")

# グラフ表示
plt.show()
```

### 〇11.2.2 系列ラベルを設定する
* 1つのグラフに複数の系列データを表示
    * それぞれの系列データが何を表しているかを示すラベル（凡例）の設定ができる
        * **`matplotlib.pyplot.legend(["ラベル名1", "ラベル名2",…])`**
        * 各系列データのプロット時（ **`matplotlib.pyplot.plot()`** ）に **label="ラベル名"** で指定すれば、 **`matplotlib.pyplot.legend()`** （引数なし）でラベル設定ができる

#### 系列ラベル（凡例）の設定例

```python
import matplotlib.pyplot as plt
import numpy as np
## Jupyter Notebookの場合必要
%matplotlib inline

## X軸データ
x = np.linspace(-10, 10)   ## -10から10まで等間隔のデータ取得（50個）
## Y軸データ1
y1 = 0.5 * x + 10
## Y軸データ2
y2 = -0.5 * x + 15

# グリッド表示設定
plt.grid(True)

# データx,y1をグラフにプロット
# キーワード引数labelで系列ラベルを設定
plt.plot(x, y1, color="#aa0000", label="Y = 0.5 * X + 10")

# データx,y2をグラフにプロット
# キーワード引数labelで系列ラベルを設定
plt.plot(x, y2, color="g", label="Y = -0.5 * X + 15")

# 系列ラベル（凡例）表示設定
plt.legend()

# グラフ表示
plt.show()
```

## ●11.3 複数のデータを可視化する②

### ○11.3.1 図の大きさを設定する
* **図：matplotlibが（複数の）グラフを描画する領域全体**
* 図の領域に関する設定
    * **`matplotlib.pyplot.figure()`** を活用する
* 図の大きさ指定
    * **`matplotlib.pyplot.figure(figsize=(横の大きさ, 縦の大きさ))`**
        * 大きさの単位はインチ（1インチ＝2.54cm）

#### 図の大きさ設定の例

```python
import matplotlib.pyplot as plt
import numpy as np
## Jupyter Notebookの場合必要
%matplotlib inline

## X軸データ
x = np.linspace(-10, 10)   ## -10から10まで等間隔のデータ取得（50個）
## Y軸データ
y = 0.5 * x + 10

# 図の大きさ設定
plt.figure(figsize=(8, 6))

# グリッド表示設定
plt.grid(True)
        
# データx,yをグラフにプロット
plt.plot(x, y)

# グラフ表示
plt.show()
```

### ○11.3.2 サブプロットを作成する
* **サブプロット：図の中で、個々のグラフを描画する領域**
* サブプロットの作成
    * **図を分割するレイアウトとその中での当該サブプロットの位置を指定**する
    * 図figについて、2×3のレイアウトに分割し、上から2行目、左から2列目にサブプロットを作成する場合
        * **`fig.add_subplot(2, 3, 5)`**
            * 引数は（サブプロットの縦の数, サブプロットの横の数, 左上から数えたときの位置）
            * 戻り値は指定した領域のサブプロット

#### サブプロット作成の例

```python
import matplotlib.pyplot as plt
import numpy as np
## Jupyter Notebookの場合必要
%matplotlib inline

## X軸データ
x = np.linspace(-10, 10)   ## -10から10まで等間隔のデータ取得（50個）
## Y軸データ
y = 0.5 * x + 10

# 図の情報を持つオブジェクト(fig)取得
# 図の大きさも設定（横8インチ、縦4インチ）
fig = plt.figure(figsize=(8, 4))

# figに対してサブプロット作成
# 縦1行、横3列のレイアウトで、左から2列目にサブプロットオブジェクト作成
ax = fig.add_subplot(1, 3, 2)

# データx,yをグラフにプロット
ax.plot(x, y)

# その他の領域にサブプロットオブジェクト作成
for i in range(3):
    if i == 1:
        continue
    else:
        fig.add_subplot(1, 3, i+1)

# グラフ表示
plt.show()
```

### ○11.3.3 サブプロットのまわりの余白を調整する
* 図の中にサブプロットを配置していくと、軸の名前やタイトルが隣のサブプロットと重なる場合がある
    * サブプロットの周り余白を調整すれば回避できる
    * **`matplotlib.pyplot.subplots_adjust(wspace=横間隔を空ける割合, hspace=縦間隔を空ける割合)`**
        * 割合はグラフの縦軸、横軸が基準

#### サブプロットの周りの余白の調整の例

```python
import matplotlib.pyplot as plt
import numpy as np
## Jupyter Notebookの場合必要
%matplotlib inline

## X軸データ
x = np.linspace(-10, 10)   ## -10から10まで等間隔のデータ取得（50個）
## Y軸データ
y = 0.5 * x + 10

# 図の情報を持つオブジェクト(fig)取得
# 図の大きさも設定（横8インチ、縦4インチ）
fig = plt.figure(figsize=(8, 4))

# figに対してサブプロット作成
# 縦1行、横3列のレイアウトで、左から2列目にサブプロットオブジェクト作成
ax = fig.add_subplot(1, 3, 2)

# サブプロットの横の間隔を0.75の割合で空ける
plt.subplots_adjust(wspace = 0.75)

# データx,yをグラフにプロット
ax.plot(x, y)

# その他の領域にサブプロットオブジェクト作成
for i in range(3):
    if i == 1:
        continue
    else:
        fig.add_subplot(1, 3, i+1)

# グラフ表示
plt.show()
```

### ○11.3.4 サブプロット内のグラフの表示範囲を設定する
* サブプロット単位でグラフの表示範囲を設定する
    * サブプロットオブジェクト(**ax**)に対して設定を行う
        * X軸の範囲指定： **`ax.set_xlim([最小値, 最大値])`**
        * Y軸の範囲指定： **`ax.set_ylim([最小値, 最大値])`**
* サブプロットを用いない場合のグラフの表示範囲設定とは指定方法が異なるので注意

#### サブプロット内のグラフの表示範囲設定の例

```python
import matplotlib.pyplot as plt
import numpy as np
## Jupyter Notebookの場合必要
%matplotlib inline

## X軸データ
x = np.linspace(-10, 10)   ## -10から10まで等間隔のデータ取得（50個）
## Y軸データ
y = 0.5 * x + 10

# 図の情報を持つオブジェクト(fig)取得
# 図の大きさも設定（横8インチ、縦4インチ）
fig = plt.figure(figsize=(8, 4))

# figに対してサブプロット作成
# 縦1行、横3列のレイアウトで、左から2列目にサブプロットオブジェクト作成
ax = fig.add_subplot(1, 3, 2)

# サブプロットの横の間隔を0.75の割合で空ける
plt.subplots_adjust(wspace = 0.75)

# サブプロットaxのY軸の表示範囲を[10,20]にする
ax.set_ylim([10, 20])

# データx,yをグラフにプロット
ax.plot(x, y)

# その他の領域にサブプロットオブジェクト作成
for i in range(3):
    if i == 1:
        continue
    else:
        fig.add_subplot(1, 3, i+1)

# グラフ表示
plt.show()
```

### ○11.3.5 サブプロット内のグラフの要素に名前を設定する
* サブプロット単位でグラフの表示範囲を設定する
    * サブプロットオブジェクト(ax)に対して設定を行う
        * グラフタイトルの設定
            * **`ax.set_title("タイトル")`**
        * X軸の名前
            * **`ax.set_xlabel("X軸の名前")`**
        * Y軸の名前
            * **`ax.set_ylabel("Y軸の名前")`**
* サブプロットを用いない場合の名前の設定とは指定方法が異なるので注意

#### □問題

```python
import matplotlib.pyplot as plt
import numpy as np
## Jupyter Notebookの場合必要
%matplotlib inline

## X軸データ
x = np.linspace(-10, 10)   ## -10から10まで等間隔のデータ取得（50個）
## Y軸データ
y = 0.5 * x + 10

# 図の情報を持つオブジェクト(fig)取得
# 図の大きさも設定（横8インチ、縦4インチ）
fig = plt.figure(figsize=(8, 4))

# figに対してサブプロット作成
# 縦1行、横3列のレイアウトで、左から2列目にサブプロットオブジェクト作成
ax = fig.add_subplot(1, 3, 2)

# サブプロットの横の間隔を0.75の割合で空ける
plt.subplots_adjust(wspace = 0.75)

## グラフタイトル設定
ax.set_title("y = 0.5 * x + 10")
## X軸設定
ax.set_xlabel("X")
## Y軸設定
ax.set_ylabel("Y")

# データx,yをグラフにプロット
ax.plot(x, y)

# その他の領域にサブプロットオブジェクト作成
for i in range(3):
    if i == 1:
        continue
    else:
        fig.add_subplot(1, 3, i+1)

# グラフ表示
plt.show()
```

### ○11.3.7 サブプロット内のグラフの軸に目盛りを設定する
* サブプロット(ax)にも目盛りを手動で設定することが可能
    * 設定方法はサブプロットを用いない場合と異なるので注意する
    * X軸に挿入する目盛りの位置
        * **`ax.set_xticks(挿入位置のリスト)`**
    * X軸に挿入するラベルの情報
        * **`ax.set_xticklabels(目盛りのリスト)`**
    * Y軸に挿入する目盛りの位置
        * **`ax.set_yticks(挿入位置のリスト)`**
    * Y軸に挿入するラベルの情報
        * **`ax.set_yticklabels(目盛りのリスト)`**

#### サブプロット内のグラフの軸に目盛りを設定する例

```python
import matplotlib.pyplot as plt
import numpy as np
## Jupyter Notebookの場合必要
%matplotlib inline

## X軸データ
x = np.linspace(-10, 10)   ## -10から10まで等間隔のデータ取得（50個）
## Y軸データ
y = 0.5 * x + 10

# 図の情報を持つオブジェクト(fig)取得
# 図の大きさも設定（横8インチ、縦4インチ）
fig = plt.figure(figsize=(8, 4))

# figに対してサブプロット作成
# 縦2行、横4列のレイアウトで、左から2列目にサブプロットオブジェクト作成
ax = fig.add_subplot(1, 3, 2)

# サブプロットの横の間隔を0.75の割合で空ける
plt.subplots_adjust(wspace = 0.75)

## グラフタイトル設定
ax.set_title("y = 0.5 * x + 10")
## X軸設定
ax.set_xlabel("X")
## Y軸設定
ax.set_ylabel("Y")

## X軸の目盛り位置とラベルの指定
ax.set_xticks([-5.0, 0, 5.0])
ax.set_xticklabels(["▲5", "0", "5"])

## グリッド表示設定
ax.grid(True)

# データx,yをグラフにプロット
ax.plot(x, y)

# その他の領域にサブプロットオブジェクト作成
for i in range(3):
    if i == 1:
        continue
    else:
        fig.add_subplot(1, 3, i+1)

# グラフ表示
plt.show()
```
