## 12.1 折れ線グラフ

### 12.1.1 マーカーの種類と色を設定する
* 折れ線グラフの描画
    * **`matplotlib.pyplot.plot()`**
* マーカーの種類の設定
    * 引数に **`marker="指定子"`** を指定する
        * マーカーの種類 [matplotlib.markers](https://matplotlib.org/3.1.1/api/markers_api.html#module-matplotlib.markers)
* マーカーの色の設定
    * 引数に **`markerfacecolor="指定子"`** を指定する
        * 指定できる色は11.2.1節の色の指定と同じ

#### マーカーの種類と色の設定例

```python
import matplotlib.pyplot as plt
import numpy as np
## Jupyter Notebookの場合必要
%matplotlib inline

## X軸データ
## -10から10まで等間隔のデータ取得（10個）
x = np.linspace(-10, 10, 10)   
## Y軸データ
y = 0.5 * x + 10

# データx,yをグラフにプロット
# マーカーを四角、マーカーの色を赤に設定
plt.plot(x, y, marker="s", markerfacecolor="r")

# グラフ表示
plt.show()
```

### 12.1.2 線のスタイルと色を設定する
* 線のスタイルの設定
    * 引数に **`linestyle="指定子"`** を指定する
        * 主な線のスタイルの種類
            * **`"-"`** ：実線
            * **`"--"`** ：破線
            * **`"-."`** ：破線（点入り）
            * **`":"`** ：点線
* 線の色の設定
    * 引数に **`color="指定子"`** を指定する
        * 指定できる色は11.2.1節のプロットの色の指定と同じ

#### 線のスタイルと色の設定例

```python
import matplotlib.pyplot as plt
import numpy as np
## Jupyter Notebookの場合必要
%matplotlib inline

## X軸データ
## -10から10まで等間隔のデータ取得（10個）
x = np.linspace(-10, 10, 10)   
## Y軸データ
y = 0.5 * x + 10

# データx,yをグラフにプロット
# 線のスタイルを点線、線の色を緑に設定
# マーカーを四角、マーカーの色を赤に設定
plt.plot(x, y, linestyle=":", color="g", marker="s", markerfacecolor="r")

# グラフ表示
plt.show()
```

## 12.2 棒グラフ

### 12.2.1 棒グラフを作成する
* 棒グラフの描画
    * **`matplotlib.pyplot.bar(横軸の値, 縦軸の値)`**

#### 棒グラフ作成の例

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

## 横軸の値を持つリスト
x = [4, 5, 6, 7, 8, 9]
## 横軸の値を持つリスト
y = [138, 151, 149, 183, 189, 160]

# 棒グラフを作成
plt.bar(x, y)

plt.show()
```

### 12.2.2 横軸にラベルを設定する
* 棒グラフの横軸にラベルを付ける方法
    * 折れ線グラフやほかのグラフでの方法と異なる
    * **`matplotlib.pyplot.bar()`** に引数 **`tick_label`** としてラベルが入っているリストを渡す

#### 横軸にラベルを設定した例

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

## 横軸の値を持つリスト
x = [1, 2, 3, 4, 5]
## 縦軸の値を持つリスト
y = [312, 308, 247, 103, 65]
## 横軸のラベルを持つリスト
labels = ["Koria", "China", "Taiwan", "HongKong", "Thai"]

# 棒グラフを作成
plt.bar(x, y, tick_label=labels)

plt.show()
```

### 12.2.3 積み上げ棒グラフを作成する
* 積み上げ棒グラフ
    * 2系統以上のデータを同じ項目に関して積み上げたグラフ
    * **`matplotlib.pyplot.bar()`** に引数 **`bottom`** として、下側のデータ列を渡す
* bottomによる設定を繰り返すことで積み上げを重ねることができる
* **`pyplot.legend(("系統1のラベル", "系統2のラベル", ...))`** とすることで、系統ラベル（凡例）の設定が可能

#### 積み上げ棒グラフの作成例

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

## 横軸の値を持つリスト
x = [1, 2, 3, 4, 5, 6]
## 縦軸の値を持つリストその1
y1 = [749, 1017, 1149, 1028, 368, 342]
## 縦軸の値を持つリストその2
y2 = [773, 912, 843, 419, 668, 601]
## 縦軸の値を持つリストその3
y3 = [100, 200, 150, 200, 300, 500]

# 積み上げ棒グラフを作成
plt.bar(x, y1)
# y1の値の上にy2の値を積み上げる
plt.bar(x, y2, bottom=y1)

# y1の値の上にy2の値を積み上げたy12の上にy3の値を積み上げる
y12 = np.array(y1) + np.array(y2) 
plt.bar(x, y3, bottom=y12)

# 凡例を作成することも可能
plt.legend(("y1", "y2", "y3"))
plt.show()
```

## 12.3 ヒストグラム

### 12.3.1 ヒストグラムを作成する
* ヒストグラムの作成
    * **`matplotlib.pyplot.hist()`**

#### ヒストグラムの作成例

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(0)
## 確率0.5で成功する試行を100回行った時の成功数のカウントを
## 1000回行った結果をdataに格納
data = np.random.binomial(100,0.5,size=1000)

# dataを用いてヒストグラムを作成
plt.hist(data)

plt.show()
```

### 12.3.2 ビン数を設定する
* ヒストグラムの階級数の設定
    * **`matplotlib.pyplot.hist()`** の引数 **`bins`**で指定する
    * **`bins="auto"`** とすると自動で設定される

#### □問題

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(0)
## 確率0.5で成功する試行を100回行った時の成功数のカウントを
## 1000回行った結果をdataに格納
data = np.random.binomial(100,0.5,size=1000)

# dataを用いてヒストグラムを作成
# 引数binsで階級数を10に指定
plt.hist(data, bins=20)

plt.show()
```

### 12.3.3 正規化を行う
* ヒストグラムの正規化
    * ヒストグラムの分布を正規分布と仮定し、合計値が1となるように操作すること
    * **`hist()`** の引数に **`density=True`** を渡す
        * 教科書では **normed** が記載されているが、現在のバージョンでは非推奨となっているので注意

#### 正規化の例

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(0)
## 確率0.5で成功する試行を100回行った時の成功数のカウントを
## 1000回行った結果をdataに格納
data = np.random.binomial(100,0.5,size=1000)

# dataを用いてヒストグラムを作成
# 引数binsで階級数を10に指定
# 正規化を行う
plt.hist(data, bins=10, density=True)

plt.show()
```

### 12.3.4 累積ヒストグラムを作成する
* 累積ヒストグラム
    * 最低階級から階級が上がるごとに数を積み上げていくヒストグラム
    * **`hist()`** の引数に **`cumulative=True`** を渡す

#### 累積ヒストグラムの例

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(0)
## 確率0.5で成功する試行を100回行った時の成功数のカウントを
## 1000回行った結果をdataに格納
data = np.random.binomial(100,0.5,size=1000)

# dataを用いてヒストグラムを作成
# 引数binsで階級数を10に指定
# 累積ヒストグラムとする
plt.hist(data, bins=10, cumulative=True)

plt.show()
```

## 12.5 円グラフ

### 12.5.1 円グラフを作成する
* 円グラフの作成
    * **`matplotlib.pyplot.pie()`**
* **`matplotlib.pyplot.axis("equal")`** を指定することで真円のグラフとなる
    * 指定しないと楕円のグラフとなる

#### 円グラフの作成例

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = [5, 10, 15, 30, 40]

# dataを円グラフとして可視化
plt.pie(data)

# 円グラフが真円になるように指定
plt.axis("equal")

plt.show()
```

### 12.5.2 円グラフにラベルを設定する
* 円グラフの各データにラベルの設定が可能
    * **`pie()`** の引数で **`labels="ラベル用のリスト"`** として渡す

#### ラベル設定の例

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = [5, 10, 15, 30, 40]
cities = ["Sapporo", "Hakodate", "Asahikawa", "Obihiro", "Kushiro"]

# dataを円グラフとして可視化
# グラフラベルも設定
plt.pie(data, labels=cities)

# 円グラフが真円になるように指定
plt.axis("equal")

plt.show()
```

### 12.5.3 特定の要素を目立たせる
* グラフ中の特定の要素だけを切り離して強調させることが可能
    * **`pie()`** の引数で **`explode="目立たせ度合いのリスト"`** として渡す
        * 目立たせ度合いは0から1の値で指定する

#### 特定の要素を目立たせる例

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = [5, 10, 15, 30, 40]
cities = ["Sapporo", "Hakodate", "Asahikawa", "Obihiro", "Kushiro"]
exp = [0, 0, 0.1, 0.2, 0.3]

# dataを円グラフとして可視化
# グラフラベルも設定
# explode指定で要素を目立たせる
plt.pie(data, labels=cities, explode=exp)

# 円グラフが真円になるように指定
plt.axis("equal")

plt.show()
```

## 12.6 3Dグラフ

### 12.6.1 3D Axesを作成する
* 3Dグラフを描画する
    * 3D描画機能を持ったサブプロットを作成する
    * **`matplotlib(.pyplot).figure().add_subplot(1,1,1,projection="3d")`**
        * **`add_subplot()`** の引数に **`projection="3d"`** と指定する
    * **mpl_toolkits.mplot3d** の **Axes3D** をimportする必要がある
    * 3Dの曲面を描画（**`plot_surface()`**）する場合、各軸のデータは2次元配列で用意する
        * **`numpy.meshgrid()`** 関数を用いるとX軸・Y軸の格子状のデータを作成しやすい

#### 3Dグラフ描画の例

```python
import numpy as np
import matplotlib.pyplot as plt
# 3D描画を行うために必要なライブラリ
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

# X軸データ
x = np.linspace(-1, 1, 50)
# Y軸データ
y = np.linspace(-1, 1, 50)
# X軸・Y軸データを組み合わせた格子状のデータを作成
X, Y = np.meshgrid(x, y)
# Z軸データ
Z = np.cos(X)

# Figureオブジェクトを作成
fig = plt.figure(figsize=(6,6))
# 3D描画できるサブプロットaxを作成
ax = fig.add_subplot(1,1,1,projection="3d")

# 曲面をプロット
ax.plot_surface(X, Y, Z)
plt.show()
```

### 12.6.2 曲面を作成する
* サブプロット **ax** に3次元のデータを利用して曲面を描画する
    * **`ax.plot_surface(X, Y, Z)`**
    * X,Y,Zデータは2次元配列とする

#### 曲面作成の例

```python
import numpy as np
import matplotlib.pyplot as plt
# 3D描画を行うために必要なライブラリ
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

# X軸データ
x = np.linspace(-1, 1, 50)
# Y軸データ
y = np.linspace(-1, 1, 50)
# X軸・Y軸データを組み合わせた格子状のデータを作成
X, Y = np.meshgrid(x, y)
# Z軸データ
Z = np.cos(X**2 - Y**2)

# Figureオブジェクトを作成
fig = plt.figure(figsize=(6,6))
# 3D描画できるサブプロットaxを作成
ax = fig.add_subplot(1,1,1,projection="3d")

# 曲面をプロット
ax.plot_surface(X, Y, Z)
plt.show()
```

### 12.6.3 3Dヒストグラムを作成する
* サブプロット **ax** に3次元でヒストグラムまたは棒グラフを作成する
    * **`ax.bar3d(xpos, ypos, zpos, dx, dy, dz)`**
        * xpos, ypos, zpos：各軸の座標
        * dx, dy, dz：各軸方向の変化量（グラフの高さ）

#### 3Dヒストグラムの例

```python
import matplotlib.pyplot as plt
import numpy as np
# 3D描画を行うために必要なライブラリ
from mpl_toolkits.mplot3d import Axes3D
import random
%matplotlib inline

np.random.seed(0)
random.seed(0)
# X,Yの座標用データ
l = set(np.arange(0,10))

# Figureオブジェクトを作成
fig = plt.figure(figsize=(5, 5))
# サブプロットaxを作成
ax = fig.add_subplot(1, 1, 1, projection="3d")
# x, y, zの座標データ
xpos = random.sample(l, 10)
ypos = random.sample(l, 10)
zpos = np.zeros(10)
# x, y, zの増加量
dx = np.ones(10)
dy = np.ones(10)
dz = np.random.choice(np.arange(1,10), 10)

# 3次元のヒストグラムを作成してください
ax.bar3d(xpos, ypos, zpos, dx, dy, dz)

plt.show()
```

### 12.6.4 3D散布図を作成する
* サブプロット **axに** 3種類のデータを用いた3次元散布図の作成
    * **`ax.scatter3d(X, Y, Z)`**
* 多次元データを1次元に変換する場合、 **`numpy.ravel()`** が利用できる

#### 3次元散布図の例

```python
import numpy as np
import matplotlib.pyplot as plt
# 3D描画を行うために必要なライブラリ
from mpl_toolkits.mplot3d import Axes3D
np.random.seed(0)
%matplotlib inline

# X, Y, Z軸のデータの作成
X = np.random.choice(np.arange(1,100), 500)
Y = np.random.choice(np.arange(1,100), 500)
Z = np.random.choice(np.arange(1,100), 500)

# Figureオブジェクトを作成
fig = plt.figure(figsize=(6, 6))
# サブプロットaxを作成
ax  = fig.add_subplot(1, 1, 1, projection="3d")
# 3D散布図を作成
ax.scatter3D(X, Y, Z)

plt.show()
```

### 12.6.5 3Dグラフにカラーマップを適用する
* 3Dグラフの第3引数（Z軸方向のデータ）についてカラーマップを適用できる
    * **matplotlib** から **cm** をimport
    * プロット時に引数として **`cmap=cm.coolwarm`** を渡す
    * プロットに対して **`colorbar()`** 関数を利用することでカラーバーの表示もできる

#### 3Dグラフにカラーマップを適用した例

```python
import numpy as np
import matplotlib.pyplot as plt
# 3D描画を行うために必要なライブラリ
from mpl_toolkits.mplot3d import Axes3D
# カラーマップを表示するためのライブラリ
from matplotlib import cm
%matplotlib inline

# X軸データ
x = np.linspace(-1, 1, 50)
# Y軸データ
y = np.linspace(-1, 1, 50)
# X軸・Y軸データを組み合わせた格子状のデータを作成
X, Y = np.meshgrid(x, y)
# Z軸データ
Z = np.cos(X**2 - Y**2)

# Figureオブジェクトを作成
fig = plt.figure(figsize=(7,7))
# 3D描画できるサブプロットaxを作成
ax = fig.add_subplot(1,1,1,projection="3d")

# 曲面をプロット
# Z軸データについてカラーマップを適用
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)

# カラーバーの表示も可能
fig.colorbar(surf)

plt.show()
```
