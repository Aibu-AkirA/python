## Seleniumのインストール

　Anacondaのコマンドプロンプトで以下の方法でpipインストール

```
pip install selenium
```

## Google Chrome のChromeDriverをインストール

　Anacondaのコマンドプロンプトで以下の方法でpipインストール

```
pip install chromedriver_binary==chromedriverのバージョン
```

chromedriverのバージョンはこちらでしらべよよう

https://sites.google.com/a/chromium.org/chromedriver/downloads

私はこちらのバージョンでやりました

```
pip install chromedriver_binary==88.0.4324.96
```
