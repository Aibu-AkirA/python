## 元々の目標

#### 弦の音を認識してリアルタイムで弦のコードと㎐を表示するプログラムをDjangoを用いて作りなさい。

# 最終結果

自動で今優先にしているシステムのマイクを探して0.2秒づつ録音します。録音したデータから音の大きさを出力します。そして、データをwave化をして時間別の振幅を表します。wave化したデータはfft(フーリエ変換)を用いて平均化したデータを出力する事までできました。

# 失敗の理由

## 1. うぬぼれと情報蒐集
<br>
一番大きいかった原因だと思います。音声のデータも画像をデータ化してlistに格納することと同じように人からlistの中身を読んで理解して使うことができると思いました。ただ、マイクの操作だけ探してデータの処理をすると何とかできそうだと思って真剣に事前調査もしなくて、前半的な計画と、他の機能を追加することだけ気にしました。しかし、システムからマイクを探して使うことも簡単にできなかったし、ただlistに格納したデータもそのままでは読められませんでした。このように自分の得意ではないことをちゃんと調べてみないて一人で挑戦したうぬぼれもひとつの大きい原因だと思います。


## 2. 音声に関する理解と基本知識の不足
<br><br>
音声に関する理解と基本知識が足りなかったことは自分が探したコードの動き方と原理にを見ることでも多くの影響をあげました。<br>
例えばマイクを操作する前に決めてあげるべきな数値が<br>

```python

CHANNELS = 1

RATE = 44100

CHUNK = 1024

RECORD_SECONDS = 0.2

```

<br>
このように何個ありますが、この数値の一つ一つが理解できなくて多くの困難を経験しました。<br>
その後のデータ処理の部分でもwave化する理由と、平均化するためにフーリエ変換する必要があるなどは短期間で理解するには難しい問題でした。
基本知識が足りなかった私はただ、コード一行で音声のデータの㎐が出て、<br>
<img src=".\img\CODE.jpg" width="700" alt="コード">
ギターのコードを配列に入れて比べて見ると簡単にできると思いました。

## 3. 時間管理と積極的な質問

自分の時間管理不足でたくさんの時間を投資できませんでした。そして、先生からフーリエ変換について助けられた後からは半分以上進んだと思って、後もうちょっと頑張ればいい結果が出ると思いましたが、その時に自分の力でやりたいと思ったので、先生が「音関係は落とし穴が大量にあります。私と逐一相談しながら進めるといいかと思います」と助言をしてくれたが積極的に質問をしかった原因も少なくないと思います。
