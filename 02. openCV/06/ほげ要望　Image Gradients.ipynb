# Image Gradients

Gradient(기울기)는 스칼라장(즉, 공간)에서 최대의 증가율을 나타내는 벡터장(방향과 힘)을 뜻합니다.<br>
영상처리에서 gradient는 영상의 edge 및 그 방향을 찾는 용도로 활용이 되는데요.<br>
이미지 (x,y)에서의 벡터값(크기와 방향, 즉 밝기와 밝기의 변화하는 방향)을 구해서 해당 pixel이 edge에 얼마나 가까운지,<br>
그리고 그 방향이 어디인지 쉽게 알수 있게 합니다. (일반적으로 이미지의 Gradient를 생각해보시면 밝기의 변화와 그 방향을 알 수 있습니다.)<br><br>
radient（傾ける）はスカラー場（即ち、空間）において最大の増加率を示すベクトル場（方向と力）を意味します。<br>
映像処理におけるgradientは、映像のedgeやその方向性を探す用途として活用されています。<br>
画像（x、y）におけるベクトル値（大きさと方向、即ち、明るさと明るさの変化する方向）を求め、そのpixelがedgeにどれだけ近いか、<br>
そしてその方向がどこであるかを容易に知ることができます。<br>
（一般的にイメージのグレードアップを考えると、明るさの変化とその方向がわかります。）<br><br>

## Sobel & Scharr Filter
Gaussian smoothing과 미분을 이용한 방법입니다. 그래서 노이즈가 있는 이미지에 적용하면 좋습니다.<br>
X축과 Y축을 미분하는 방법으로 경계값을 계산합니다.<br><br>
직선을 미분하면 상수, 곡선을 미분하면 또 다른 방정식이 나오는 성질을 이용하여 edge에 대한 선을 그려주는 기능을 합니다.<br>
X축 미분은 수평선(수직선이 남음), Y축 미분은 수직선(수평선이 남음)을 미분하여 경계가 사라지는 효과가 있습니다.<br>
미분시 소실되는 표본의 정보가 많을 수 있어 aperture_size 값을 이용하여 소실되는 정도를 조절할 수 있습니다.<br><br>
만약 ksize가 -1이면 3x3 Scharr filter가 적용이 되어 Sobel의 3x3보다 좀 더 나은 결과를 보여 줍니다.<br><br>


Gaussian smoothingと微分を利用した方法です。 だからノイズがあるイメージに適用するといいです。<br>
X軸とY軸を微分する方法によりしきい値を計算します。<br><br>
直線を微分すると定数、曲線を微分するとまた別の方程式が出てくる性質を利用し、edgeに対する線を描く機能をします。
<br>X軸微分は水平線(垂直線が残る)、Y軸微分は垂直線(水平線が残る)を微分することで境界が消える効果があります。<br>
未分時に消失する標本の情報が多いため、 aperture_size 値を利用して消失する程度を調節できます。<br><br>
もしksizeが-1なら、3x3 Scharrfilterが適用され、Sobelの3x3より良い結果を示します。<br><br>

#### cv2.Sobel(src, ddepth, dx, dy[, dst[, ksize[, scale[, delta[, borderType]]]]]) → dst<br>

src – input image<br>
ddepth – output image의 depth, -1なら input imageと同じ.<br>
dx – x軸の微分の次数.<br>
dy – y軸の微分の次数.<br>
ksize – kernel size(ksize x ksize)<br>
cv2.Scharr(src, ddepth, dx, dy[, dst[, scale[, delta[, borderType]]]]) → dst<br>
cv2.Sobel() と同じだけどksizeがsobelの3x3より正確に適用されます。<br>

## Laplacian
이미지의 가로와 세로에 대한 Gradient를 2차 미분한 값입니다.<br>
Sobel filter에 미분의 정도가 더해진 것과 비슷합니다.<br>
(dx와 dy가 2인 경우) blob(주위의 pixel과 확연한 picel차이를 나타내는 덩어리)검출에 많이 사용됩니다.<br><br>
画像の横と縦に対するGradientを二次微分した値です。<br>
Sobel filterに微分の程度が加わったのと似ています(dxとdyが2の場合)。<br>
blob（周囲のpixelと明らかなpicel差を示す塊）の検出に多く用いられます。<br>

#### v2.Laplacian(src, ddepth[, dst[, ksize[, scale[, delta[, borderType]]]]]) → dst<br>
src – source image<br>
ddepth – output iamgeの depth.<br>


### Canny Edge Detection
가장 유명한 Edge Detection방법입니다. 여러 단계의 Algorithm을 통해서 경계를 찾아 냅니다.<br>
1. Noise Reduction<br>
이미지의 Noise를 제거합니다. 이때 5x5의 Gaussian filter를 이용합니다.<br>
2. Edge Gradient Detection<br>
이미지에서 Gradient의 방향과 강도를 확인합니다. 경계값에서는 주변과 색이 다르기 때문에 미분값이 급속도로 변하게 됩니다.<br>
이를 통해 경계값 후보군을 선별합니다.<br>
3. Non-maximum Suppression<br>
이미지의 pixel을 Full scan하여 Edge가 아닌 pixel은 제거합니다.<br>
4. Hysteresis Thresholding<br>
이제 지금까지 Edge로 판단된 pixel이 진짜 edge인지 판별하는 작업을 합니다.<br>
max val과 minVal(임계값)을 설정하여 maxVal 이상은 강한 Edge, min과 max사이는 약한 edge로 설정합니다.<br>
이제 약한 edge가 진짜 edge인지 확인하기 위해서 강한 edge와 연결이 되어 있으면 edge로 판단하고, 그러지 않으면 제거합니다.<br>
이와 같은 일련의 작업을 통해서 경계값만을 남겨두고 제거합니다.<br>

最も有名なEdge Detection方法です。 複数段階のAlgorithmを通じて境界を見つけます。<br>
1. Noise Reduction<br>
画像のNoiseを除去します。 この時、5x5のGaussian filterを利用します。<br>
2. Edge Gradient Detection<br>
画像でGradientの方向と強度を確認します。 しきい値では周囲と色が異なるため、微分値が急速に変化します。<br>
これによりしきい値候補群を選別します。<br>
3. Non-maximum Suppression<br>
画像のpixelをFullscanしてEdgeではないpixelは削除します。<br>
4. Hysteresis Thresholding<br>
次に、ここまでEdgeで判断されたpixelが本当にedgeなのかを判別する作業をします。<br>
maxvalとminVal（しきい値）を設定し、maxVal以上は強いEdge、minとmax間は弱いedgeに設定します。<br>
この後、弱いedgeが本当にedgeなのかを確認するため、強いedgeと結びついていればedgeで判断して、そうでなければ取り除きます。<br>
このような一連の作業を通じて、しきい値だけを残して除去します。<br>

#### cv2.Canny(image, threshold1, threshold2[, edges[, apertureSize[, L2gradient]]]) → edges

image – 8-bit input image<br>
threshold1 – Hysteresis Thredsholding の min の値<br>
threshold2 – Hysteresis Thredsholding の max の値<br>


```python
#-*- coding:utf-8 -*-
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('dave.jpg')
canny = cv2.Canny(img,30,70)

laplacian = cv2.Laplacian(img,cv2.CV_8U)
sobelx = cv2.Sobel(img,cv2.CV_8U,1,0,ksize=3)
sobely = cv2.Sobel(img,cv2.CV_8U,0,1,ksize=3)

images = [img,laplacian, sobelx, sobely, canny]
titles = ['Origianl', 'Laplacian', 'Sobel X', 'Sobel Y','Canny']

for i in range(5):
    plt.subplot(2,3,i+1),plt.imshow(images[i]),plt.title([titles[i]])
    plt.xticks([]),plt.yticks([])

plt.show()
```
