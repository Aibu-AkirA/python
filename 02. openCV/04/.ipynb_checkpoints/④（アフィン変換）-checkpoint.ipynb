## フリップ

```python
# -*- coding: utf-8 -*

import cv2

try:
    img = cv2.imread('./lena.png')

    if img is None:
        print ('ファイルを読み込めません。')
        import sys
        sys.exit()

    dst = cv2.flip(img, 0)
    cv2.imwrite('./flip0.png', dst)
    cv2.imshow('./flip0.png', dst)

    dst = cv2.flip(img, 1)
    cv2.imwrite('./flip1.png', dst)
    cv2.imshow('./flip1.png', dst)

    dst = cv2.flip(img, -1)
    cv2.imwrite('./flip-1.png', dst)
    cv2.imshow('./flip-1.png', dst)

except:
    import sys
    print("Error:", sys.exc_info()[0])
    print(sys.exc_info()[1])
    import traceback
    print(traceback.format_tb(sys.exc_info()[2]))
    
#画面の削除
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## リサイズ

```python
# -*- coding: utf-8 -*

import cv2

try:
    img = cv2.imread('./lena.png')

    if img is None:
        print ('ファイルを読み込めません。')
        import sys
        sys.exit()

    SCALE1 = 0.5
    SCALE2 = 1.62
    height = img.shape[0]
    width = img.shape[1]

    dst = cv2.resize(img, (int(width*SCALE1), int(height*SCALE1)))
    cv2.imwrite('./resize0.5.png', dst)
    cv2.imshow('./resize0.5.png', dst)

    dst = cv2.resize(img, (int(width*SCALE2), int(height*SCALE2)))
    cv2.imwrite('./resize1.62.png', dst)
    cv2.imshow('./resize1.62.png', dst)

    dst = cv2.resize(img, (400, 200))
    cv2.imwrite('./resize400x200.png', dst)
    cv2.imshow('./resize400x200.png', dst)

except:
    import sys
    print("Error:", sys.exc_info()[0])
    print(sys.exc_info()[1])
    import traceback
    print(traceback.format_tb(sys.exc_info()[2]))
    
#画面の削除
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 回転

```python
# -*- coding: utf-8 -*

import cv2

try:
    img = cv2.imread('./lena.png')

    if img is None:
        print ('ファイルを読み込めません。')
        import sys
        sys.exit()

    height = img.shape[0]
    width = img.shape[1]
    center = (int(width/2), int(height/2))

    affin_trans = cv2.getRotationMatrix2D(center, 33.0, 1.0)
    dst = cv2.warpAffine(img, affin_trans, (width, height))
    cv2.imwrite('./rotate_033.png', dst)
    cv2.imshow('./rotate_033.png', dst)

    affin_trans = cv2.getRotationMatrix2D(center, 110.0, 1.2)
    dst = cv2.warpAffine(img, affin_trans, (width, height), flags=cv2.INTER_CUBIC)
    cv2.imwrite('./rotate_110.png', dst)
    cv2.imshow('./rotate_110.png', dst)

except:
    import sys
    print("Error:", sys.exc_info()[0])
    print(sys.exc_info()[1])
    import traceback
    print(traceback.format_tb(sys.exc_info()[2]))
    
#画面の削除
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 透視投影

```python
# -*- coding: utf-8 -*

import cv2
import numpy as np

try:
    img = cv2.imread('./lena.png')

    if img is None:
        print ('ファイルを読み込めません。')
        import sys
        sys.exit()

    rows, cols = img.shape[:2]
    x0 = cols/4
    x1 = (cols*3)/4
    y0 = rows/4
    y1 = (rows*3)/4

    list_srcs = np.float32([[x0, y0],[x0, y1],[x1, y1],[x1, y0]])

    #pattern-0
    x_margin = cols/10
    y_margin = rows/10
    list_dsts = np.float32([[x0+x_margin, y0+y_margin],list_srcs[1],list_srcs[2],[x1-x_margin, y0+y_margin]])

    perspective_matrix = cv2.getPerspectiveTransform(list_srcs,list_dsts)
    dst = cv2.warpPerspective(img, perspective_matrix, (cols, rows))

    cv2.imwrite("./dst0.png", dst)
    cv2.imshow("./dst0.png", dst)

    #pattern-1
    x_margin = cols/8
    y_margin = rows/8
    list_dsts = np.float32([list_srcs[0],list_srcs[1],[x1-x_margin, y1-y_margin],[x1-x_margin, y0+y_margin]])

    perspective_matrix = cv2.getPerspectiveTransform(list_srcs,list_dsts)
    dst = cv2.warpPerspective(img, perspective_matrix, (cols, rows))

    cv2.imwrite("./dst1.png", dst)
    cv2.imshow("./dst1.png", dst)
    
    #pattern-2
    x_margin = cols/6
    y_margin = rows/6
    list_dsts = np.float32([[x0+x_margin, y0+y_margin],list_srcs[1],[x1-x_margin, y1-y_margin],list_srcs[3]])

    perspective_matrix = cv2.getPerspectiveTransform(list_srcs,list_dsts)
    dst = cv2.warpPerspective(img, perspective_matrix, (cols, rows))

    cv2.imwrite("./dst2.png", dst)
    cv2.imshow("./dst2.png", dst)
    
except:
    import sys
    print("Error:", sys.exc_info()[0])
    print(sys.exc_info()[1])
    import traceback
    print(traceback.format_tb(sys.exc_info()[2]))
    
#画面の削除
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## トリミング

```python
# -*- coding: utf-8 -*

import cv2

try:
    img = cv2.imread('./lena.png')

    if img is None:
        print ('ファイルを読み込めません。')
        import sys
        sys.exit()

    height = img.shape[0]
    width = img.shape[1]

    dst = img[40:height, 40:width]
    cv2.imwrite('./Trimming.jpg', dst)
    cv2.imshow('./Trimming.jpg', dst)

except:
    import sys
    print("Error:", sys.exc_info()[0])
    print(sys.exc_info()[1])
    import traceback
    print(traceback.format_tb(sys.exc_info()[2]))
    
#画面の削除
cv2.waitKey(0)
cv2.destroyAllWindows()
```
