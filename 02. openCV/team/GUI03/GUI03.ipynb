## フィルターのみの処理（未完成）

現状、青いフィルターが適用された映像を１フレームだけ出力してエラーで落ちる

エラー内容は"映像を格納した配列"と"青いフィルターを格納した配列"のサイズ不一致

```python
# -*- coding: utf-8 -*

import cv2
import numpy as np

try:
    # ビデオをキャプチャー.
    cap = cv2.VideoCapture(0)
    
    while(1):
        # ビデオの各フレームを撮ります
        _, frame = cap.read()

        if cap is None:
            print ('ファイルを読み込めません。')
            import sys
            sys.exit()

        # int～は自動で縦横の長さを取得する
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        # 青色を指定
        blue = np.zeros((height, width, 3), np.uint8)
        blue[:,:] = [128, 0, 0]
        
        # 青色を加算
        dst = cv2.add(frame, blue)
        #dst = cv2.bitwise_and(frame,blue)
        #dst = cv2.addWeighted(src1=frame, alpha=0.5,src2=blue, beta=0.5,gamma=0)
        
        # 表示
        cv2.imshow('frame',dst)
        
        #27(esc)キーでループを抜け出す処理(×を押してもループは継続する)
        k = cv2.waitKey(5) & 0xFF
        if k == 27:
            break

        #カメラを閉じる        
        cap.release()
        #ウィンドウ閉じる
        cv2.destroyAllWindows()

except:
    import sys
    print("Error:", sys.exc_info()[0])
    print(sys.exc_info()[1])
    import traceback
    print(traceback.format_tb(sys.exc_info()[2]))
```
